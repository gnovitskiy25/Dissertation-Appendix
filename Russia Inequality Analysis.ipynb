#### Plotting - Income Inequality - Over GDP



Think about potential error of the mean for all the plots.

```python
ax = sns.kdeplot(income_inequality['pretaxincome_over_total_pretax_income'] , shade = True)
plt.axvline(income_inequality['pretaxincome_over_total_pretax_income'].mean(), linestyle = "--" , color = "blue")
plt.axvline(income_inequality['pretaxincome_over_total_pretax_income'].median(), linestyle = "--" , color = "darkblue")

vals = ax.get_xticks()
ax.set_xticklabels(['{:20,.2f}'.format(x) for x in vals]) 
valsy = ax.get_yticks()
ax.set_yticklabels(['{:20,.2f}'.format(x) for x in valsy]) 

plt.legend(loc='best', bbox_to_anchor=(1.3, 0.5, 0.5, 0.5))

plt.show()
```

```python
mean_p99 = income_inequality[income_inequality["Percentile"] == "p99p100"].iloc[1,3]
mean_p90 = income_inequality[income_inequality["Percentile"] == "p90p100"].iloc[1,3]
mean_p50 = income_inequality[income_inequality["Percentile"] == "50p90p"].iloc[1,3]
mean_p0 = income_inequality[income_inequality["Percentile"] == "0p50p"].iloc[1,3]
```

```python
g = sns.FacetGrid(income_inequality,  hue = "Percentile", size=4.5, aspect=2)
g.map(plt.axvline, x = mean_p99, ls=":", color = "red")
g.map(plt.axvline, x = mean_p90, ls=":", color = "lightblue" , linewidth = 2.5)
g.map(plt.axvline, x = mean_p50, ls=":", color = "yellow")
g.map(plt.axvline, x = mean_p0, ls=":", color = "green")
g.map(sns.kdeplot, "pretaxincome_over_total_pretax_income" , shade = True)


g.add_legend()
plt.show();
```

```python
inc_ine = income_inequality.copy()
inc_ine["mean"] = inc_ine["pretaxincome_over_total_pretax_income"].mean()
inc_ine.head()
```

```python
g = sns.FacetGrid(inc_ine,  col = "Percentile", legend_out = False, size= 4)
g.map(sns.kdeplot, "pretaxincome_over_total_pretax_income" , shade = True)
plt.xticks(np.arange(0, 1, step=0.15))
plt.show()
```

```python
sns.set(style='ticks')
ax = income_inequality_2.plot(kind = "line" , x = "Year" ,xlim = (1960,2014) ,  
                         linestyle='-', 
                         linewidth=2.5, 
                         drawstyle ="steps-post", 
                         marker = "." ,
                         markersize=10)

plt.axhline(mean_p99, color = "purple", linestyle = "dotted" , linewidth = 1)
plt.axhline(mean_p90, color = "blue", linestyle = "dotted" , linewidth = 1)
plt.axhline(mean_p50, color = "green", linestyle = "dotted" , linewidth = 1)
plt.axhline(mean_p0, color = "red", linestyle = "dotted" , linewidth = 1)

vals = ax.get_yticks()
ax.set_yticklabels(['{:,.2%}'.format(x) for x in vals])


ax.legend(loc='best', bbox_to_anchor=(1, 0.5, 0.5, 0.5))
```
